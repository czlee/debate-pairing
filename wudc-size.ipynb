```julia
using PyPlot
using Interact
include("powerpairing.jl")
ntournaments = 300;
```

```julia
function simulateandplot(nteams, nrounds)
    positions = simulatemanytournaments(pairpower4, ntournaments, nteams, nrounds) # in powerpairing.jl
    boxplot(positions)
    plot([0, nteams], [0, nteams], color=(1, 1, 0))
    xlabel("true rank")
    ylabel("actual rank")
    xticks(yticks()[1], map(Int, yticks()[1]))
    title("Rank distributions after $nrounds rounds")
end;
```

```julia
function zoomtopdos()
    xlim(0, 80)
    ylim(0, 80)
    plot(xlim(), [48.5, 48.5], color=(0, 1, 0))
    plot([48.5, 48.5], ylim(), color=(0, 1, 0))
    xticks(yticks()[1], map(Int, yticks()[1]))
end;
```

```julia
simulateandplot(400, 9)
zoomtopdos();
```

```julia
simulateandplot(400, 10)
zoomtopdos();
```
